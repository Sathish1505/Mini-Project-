```python
x=df2[['DT','RHOB','NPHI','GR','RT']]    #x is features
y=df2['DTS']                             #y is target
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_new=scaler.fit_transform(x)
```

```python
x_new.reshape(-1,1)
```

```python
x_new=pd.DataFrame(x_new,columns=['DT','RHOB','NPHI','GR','RT'])
x_new
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_new,y)
```

```python
from sklearn.linear_model import LinearRegression

lmodel=LinearRegression()
```

```python
lmodel.fit(xtrain,ytrain)
```

```python
ytrainp=lmodel.predict(xtrain)
ytestp=lmodel.predict(xtest)
```

```python
print(abs(ytrain-ytrainp).mean())
print(abs(ytest-ytestp).mean())
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(ytest,ytestp)
r2=r2_score(ytest,ytestp)

print(f"mean squared error : {mse:.2f}")
print(f"R^2 score : {r2:.2f}")
```

```python
plt.scatter(ytest,ytestp,c='r',s=3,label="Predicted vs Actual")
plt.plot([ytest.min(),ytest.max()],[ytest.min(),ytest.max()],c='black',lw=1,label='Prefect fit')
plt.xlabel('Actual values')
plt.ylabel('Preicted Values')
plt.legend()
plt.show()
```

```python
residuals =ytest-ytestp

plt.scatter(ytestp,residuals,c='b',s=5)

plt.axhline(0,c='r',linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Vs Predicted Values')
plt.show()
```

```python
Depth=df2['DEPTH'].iloc[xtest.index]
```

```python
plt.figure(figsize=(10,8))
plt.plot(Depth,ytest,label='Actual Vs',c='r')
plt.plot(Depth,ytestp,label='Predicted Vs',c='blue')
plt.xlabel('DEPTH')
plt.ylabel('Vs')
plt.title('Actual Vs and Predicted Vs')
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
df3=pd.DataFrame({'Depth':Depth,'ytest':ytest,'ytestp':ytestp})
```

```python
df3.reset_index(drop=True)
```

```python
df3_sort=df3.sort_values(by='Depth',ascending=True)
df3_sort=df3_sort.reset_index(drop=True)
```

```python
plt.figure(figsize=(5,10))
plt.plot(df3_sort['ytest'],df3_sort['Depth'],label='Actual Vs',c='r',lw=1)
plt.plot(df3_sort['ytestp'],df3_sort['Depth'],label='Predicted Vs',c='blue',lw=1)
plt.ylabel('DEPTH')
plt.xlabel('Vs')
plt.title('Actual Vs and Predicted Vs')
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python

x=df2[['DT','RHOB','NPHI','GR','RT']]    #x is features
y=df2['DTS']                             #y is target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rmodel=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
rmodel.fit(xtrain,ytrain)
```

```python
ytrainp_rf=rmodel.predict(xtrain)
ytestp_rf=rmodel.predict(xtest)
```

```python
print(abs(ytrain-ytrainp_rf).mean())
print(abs(ytest-ytestp_rf).mean())
```

```python
mse_rf=mean_squared_error(ytest,ytestp_rf)
r2_rf=r2_score(ytest,ytestp_rf)

print(f'random forest mean squared error :{mse_rf}')
print(f'random forest r^2 score :{r2_rf}')
```

```python
plt.scatter(ytest,ytestp_rf,c='r',s=3,label="Predicted vs Actual")
plt.plot([ytest.min(),ytest.max()],[ytest.min(),ytest.max()],c='black',lw=1,label='Prefect fit')
plt.xlabel('Actual values')
plt.ylabel('Preicted Values')
plt.legend()
plt.show()
```
